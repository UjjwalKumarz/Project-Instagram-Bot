### Importing all the required libraries and Creating a web driver session

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from selenium.common.exceptions import TimeoutException
from selenium.common.exceptions import NoSuchElementException
```

```python
def webdriverSession(url):
    driver = webdriver.Chrome(service = Service(ChromeDriverManager().install()))
    driver.maximize_window()
    driver.get(url)
    return driver

url = 'https://www.instagram.com/'
driver = webdriverSession(url)
```

## 1). Login to your Instagram Handle
### 1.1). Submit with sample username and password

```python
def LogIn(driver, username, password):
    u_name = wait.until(EC.presence_of_element_located((By.NAME, 'username')))
    u_pass = wait.until(EC.presence_of_element_located((By.NAME, 'password')))
    u_name.send_keys(username)
    u_pass.send_keys(password)

    login = driver.find_element(By.XPATH, '// button[contains(@type,"submit")]')
    login.click()
    
    try:
        save_info = wait.until(EC.presence_of_element_located((By.XPATH, '// div[@class = "cmbtv"]/button')))
        save_info.click()
    except TimeoutException:
        pass

    # Handling Turn On Notifications window
    try:
        notification = wait.until(EC.presence_of_element_located((By.XPATH, '// button[contains(@class, "_a9_1")]')))
        notification.click()
    except TimeoutException:
        pass
    finally:
        print('You have successfully Logged In')
        
wait = WebDriverWait(driver,10)
username = 'SAMPLE USERNAME'
password = 'SAMPLE PASSWORD'
LogIn(driver, username, password)
time.sleep(2)
```

## 2). Type for “food” in search bar and print all the names of the Instagram Handles that are displayed in list after typing “food”
### 2.1). Note : Make sure to avoid printing hashtags

```python
def SearchText(driver, text):
    search_box = wait.until(EC.presence_of_element_located((By.XPATH, '// div[@class = "_aaw9"]')))
    search_box.click()
    input_k = wait.until(EC.presence_of_element_located((By.XPATH, '//nav//input')))
    input_k.send_keys(text)
    
    time.sleep(3)

def FindingHandles(driver, text):
    SearchText(driver, text)
    handle = driver.find_element(By.CLASS_NAME, '_abnx')
    handles = handle.find_elements(By.CLASS_NAME, '_aacw')

    for i in handles:
        h = i.get_attribute('innerHTML')
        if h[0]=='#':
            continue
        print(h)

    clear = driver.find_element(By.CLASS_NAME, '_9-lv')
    clear.click()
    time.sleep(2)
    return
    
text = 'food'
FindingHandles(driver, text)
```

## 3). Searching and Opening a profile using 
### 3.1). Open profile of "So Delhi"

```python
def OpenAccount(driver, profile_name):
    # Searching the profile
    SearchText(driver, profile_name)
    
    # Finding and opening the profile
    handle = driver.find_element(By.CLASS_NAME, '_abnx')
    handles = handle.find_elements(By.CLASS_NAME, '_aacy')

    for i in handles:
        h = i.get_attribute('innerHTML')
        if h == profile_name:
            try:
                i.click()
                print('You have opened the profile of', profile_name)
            except:
                print('Unable to open the profile')
            time.sleep(3)
            break
    
profile_name = 'So Delhi'
OpenAccount(driver, profile_name)
```

## 5). Like/Unlike posts

### 5.1). Liking the top 30 posts of the 'dilsefoodie'. Print message if you have already liked it.

```python
def OpenProfile(driver, profile_name):
    SearchText(driver, profile_name)
    handle = driver.find_element(By.CLASS_NAME, '_abnx')
    handles = handle.find_elements(By.CLASS_NAME, '_aacw')

    for i in handles:
        h = i.get_attribute('innerHTML')
        if h == profile_name:
            i.click()
            print('We have Successfully opened the profile of', h )
            break

# Searching "dilsefoodie" and Opening the profile
profile_name = 'dilsefoodie'
OpenProfile(driver, profile_name)

def FindPosts(driver):
    # Finding posts
    posts = driver.find_elements(By.CLASS_NAME, '_aanf')
    while len(posts)<30:
        # Scrolling the page down
        driver.execute_script('window.scrollBy(0,3000);')
        time.sleep(3)
        # Finding posts
        posts = driver.find_elements(By.CLASS_NAME, '_aanf')
    
    # Scrolling the page to top
    driver.execute_script('window.scrollBy(0,-6000);')
    time.sleep(3)
    return posts

def LikePosts(driver, n):
    posts = FindPosts(driver)
    count = 0
    while count<n:
        posts[count].click()
        time.sleep(2)
        try:
            like_btn = driver.find_element(By.XPATH, '// span[@class = "_aamw"]//div[contains(@class, "_abl_")]')
            like_btn.click()
        except:
            print('Already liked this post')
        count+=1
        driver.back()
        time.sleep(2)
    print('Liked all 30')

# Liking first 30 posts
LikePosts(driver, 30)
```

### 5.2). Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.

```python
def UnlikePosts(driver, n):
    posts = FindPosts(driver)
    count = 0
    while count<n:
        posts[count].click()
        time.sleep(2)
        try:
            like_btn = driver.find_element(By.XPATH, '// span[@class = "_aamw"]//div[contains(@class, "_abl_")]')
            print("Already unliked it")

        except:
            unlike_btn = driver.find_element(By.XPATH, '// span[@class = "_aamw"]/button/div')
            unlike_btn.click()
        count += 1
        driver.back()
        time.sleep(2)
    print('Unliked All 30')

# Unliking first 30 posts
UnlikePosts(driver, 30)
```

## 6). Extract list of followers

### 6.1). Extract the usernames of the first 500 followers of "foodtalkindia" and "sodelhi".

```python
def FindFollowers(driver, profile_name, n):
    # Opening profile page
    OpenProfile(driver, profile_name)
    
    # Clicking on followers button
    li = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, '_aa_5')))
    li[1].click()
    
    # Extracting followers
    while True:
        accounts = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'notranslate')))
        if len(accounts)<n:
            driver.execute_script('arguments[0].scrollIntoView();', accounts[-1])
            time.sleep(2)
        else:
            break

    follower = []
    for i in range(500):
        follower.append(accounts[i].find_element(By.TAG_NAME, 'span').get_attribute('innerHTML'))
    driver.back()
    driver.back()
    time.sleep(3)
    return follower

# Finding followers of 'foodtalkindia'
profile_name = 'foodtalkindia'
f_foodtalkindia = FindFollowers(driver, profile_name, 500)
print('Followers of foodtalkindia are :-')
for f in f_foodtalkindia:
    print(f)
```

```python
# Finding followers of 'sodelhi'
profile_name = 'sodelhi'
f_sodelhi = FindFollowers(driver, profile_name, 500)
print('Followers :-')
for f in f_sodelhi:
    print(f)
```

### 6.2). Now print all the followers of “foodtalkindia” that you are following but those who don’t follow you.

```python
# Extracting all of my followers
def allFollowers(driver, profile_name):
    # Opening profile page
    OpenProfile(driver, profile_name)
    
    # Clicking on followers button
    li = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, '_aa_5')))
    li[1].click()
    
    # Extracting followers
    accounts = []
    n = len(accounts)
    while True:
        accounts = wait.until(EC.presence_of_all_elements_located((By.XPATH, '//li//a/span')))
        if len(accounts)!=n:
            driver.execute_script('arguments[0].scrollIntoView();', accounts[-1])
            time.sleep(4)
            n = len(accounts)
        else:
            break

    follower = []
    for i in accounts:
        follower.append(i.get_attribute('innerHTML'))
    driver.back()
    driver.back()
    time.sleep(3)
    return follower
my_followers = allFollowers(driver, username)
```

```python
# Extracting all of my followings
def allFollowing(driver, profile_name):
    # Opening profile page
    OpenProfile(driver, profile_name)
    
    # Clicking on followers button
    li = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, '_aa_5')))
    li[2].click()
    
    # Extracting followers
    accounts = []
    n = len(accounts)
    while True:
        accounts = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'notranslate')))
        if len(accounts)!=n:
            driver.execute_script('arguments[0].scrollIntoView();', accounts[-1])
            time.sleep(4)
            n = len(accounts)
        else:
            break

    following = []
    for i in accounts:
        following.append(i.find_element(By.TAG_NAME, 'span').get_attribute('innerHTML'))
    driver.back()
    driver.back()
    time.sleep(3)
    return following
me_following = allFollowing(driver, username)
```
